```python
"""
花卉识别系统 - 单文件完整版
整合了CNN模型、数据处理、训练、评估和可视化功能
"""
import os
import tarfile
import warnings
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from tqdm import tqdm

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms

warnings.filterwarnings('ignore')
```

```python
class FlowerClassifier(nn.Module):
    """简单的CNN模型，用于花卉分类"""

    def __init__(self, num_classes=5):
        super().__init__()

        # 卷积层（提取图像特征）
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)

        # 池化层
        self.pool = nn.MaxPool2d(2, 2)

        # 丢弃层
        self.dropout = nn.Dropout(0.5)

        # 全连接层
        self.fc1 = nn.Linear(128 * 18 * 18, 512)
        self.fc2 = nn.Linear(512, num_classes)

    def forward(self, x):
        # 卷积+池化
        x = self.pool(F.relu(self.conv1(x)))
        x = self.pool(F.relu(self.conv2(x)))
        x = self.pool(F.relu(self.conv3(x)))

        # 展平
        x = x.view(-1, 128 * 18 * 18)

        # 全连接
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = self.fc2(x)

        return x

```

```python
def check_and_extract_data():
    """检查并解压数据集"""
    if not os.path.exists('flower_photos.tgz'):
        print("❌ 找不到 flower_photos.tgz 文件！")
        print("请从以下链接下载：")
        print("https://storage.googleapis.com/download.tensorflow.org/example_images/flower_photos.tgz")
        print("然后拖到项目文件夹里。")
        return False

    if not os.path.exists('flower_photos'):
        print("📦 正在解压数据集...")
        try:
            with tarfile.open('flower_photos.tgz', 'r:gz') as tar:
                tar.extractall()
            print("✅ 数据集解压完成！")
        except Exception as e:
            print(f"❌ 解压失败: {e}")
            return False

    return True
```

```python
def load_and_prepare_data(data_dir='flower_photos'):
    """加载和准备数据"""
    if not os.path.exists(data_dir):
        if not check_and_extract_data():
            return None, None, None
    
    # 数据预处理
    transform = transforms.Compose([
        transforms.Resize((150, 150)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(10),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
    
    # 加载数据集
    try:
        dataset = datasets.ImageFolder(root=data_dir, transform=transform)
    except Exception as e:
        print(f"❌ 加载数据集失败: {e}")
        return None, None, None
    
    # 划分训练集和测试集
    train_size = int(0.8 * len(dataset))
    test_size = len(dataset) - train_size
    train_dataset, test_dataset = torch.utils.data.random_split(dataset, [train_size, test_size])
    
    # 创建数据加载器
    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False)
    
    # 获取类别名称
    class_names = dataset.classes
    
    print(f"✅ 数据集加载成功！")
    print(f"   类别: {class_names}")
    print(f"   总样本数: {len(dataset)}")
    print(f"   训练集: {len(train_dataset)} 张图片")
    print(f"   测试集: {len(test_dataset)} 张图片")
    
    return train_loader, test_loader, class_names
```

```python
def visualize_samples(data_loader, class_names):
    """可视化一些样本图片"""
    images, labels = next(iter(data_loader))

    fig, axes = plt.subplots(2, 5, figsize=(15, 6))
    fig.suptitle('花卉样本展示', fontsize=16)

    for i in range(10):
        ax = axes[i // 5, i % 5]
        img = images[i].numpy().transpose((1, 2, 0))

        # 反归一化
        mean = np.array([0.485, 0.456, 0.406])
        std = np.array([0.229, 0.224, 0.225])
        img = std * img + mean
        img = np.clip(img, 0, 1)

        ax.imshow(img)
        ax.set_title(f'{class_names[labels[i]]}')
        ax.axis('off')

    plt.tight_layout()
    plt.savefig('flower_samples.png', dpi=150, bbox_inches='tight')
    print("✅ 样本图片已保存为 'flower_samples.png'")
    plt.show()
```

```python
def evaluate_model(model, test_loader, device):
    """评估模型"""
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    return 100 * correct / total
```

```python
def train_model(model, train_loader, test_loader, device, num_epochs=10):
    """训练模型"""
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    train_losses = []
    train_accs = []
    test_accs = []

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        pbar = tqdm(train_loader, desc=f'Epoch {epoch + 1}/{num_epochs}')
        for images, labels in pbar:
            images, labels = images.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

            pbar.set_postfix({
                'loss': f'{loss.item():.4f}',
                'acc': f'{100 * correct / total:.2f}%'
            })

        train_acc = 100 * correct / total
        train_losses.append(running_loss / len(train_loader))
        train_accs.append(train_acc)

        test_acc = evaluate_model(model, test_loader, device)
        test_accs.append(test_acc)

        print(f'  Epoch {epoch + 1}: 训练损失 = {train_losses[-1]:.4f}, '
              f'训练精度 = {train_acc:.2f}%, 测试精度 = {test_acc:.2f}%')

    return train_losses, train_accs, test_accs
```

```python
def visualize_training(train_losses, train_accs, test_accs):
    """可视化训练过程"""
    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    axes[0].plot(train_losses, 'b-', linewidth=2)
    axes[0].set_xlabel('Epoch')
    axes[0].set_ylabel('Loss')
    axes[0].set_title('训练损失曲线')
    axes[0].grid(True, alpha=0.3)

    axes[1].plot(train_accs, 'g-', label='训练精度', linewidth=2)
    axes[1].plot(test_accs, 'r-', label='测试精度', linewidth=2)
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Accuracy (%)')
    axes[1].set_title('训练和测试精度')
    axes[1].legend()
    axes[1].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig('training_curves.png', dpi=150, bbox_inches='tight')
    print("✅ 训练曲线已保存为 'training_curves.png'")
    plt.show()

```

```python
def predict_single_image(model, image_path, class_names, device):
    """预测单张图片"""
    if not os.path.exists(image_path):
        print(f"❌ 图片不存在: {image_path}")
        return

    # 预测时的transform
    pred_transform = transforms.Compose([
        transforms.Resize((150, 150)),
        transforms.ToTensor(),
        transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])

    try:
        image = Image.open(image_path).convert('RGB')
    except Exception as e:
        print(f"❌ 无法打开图片: {e}")
        return

    image_tensor = pred_transform(image).unsqueeze(0).to(device)

    # 预测
    model.eval()
    with torch.no_grad():
        output = model(image_tensor)
        probabilities = F.softmax(output, dim=1)
        confidence, predicted = torch.max(probabilities, 1)

    # 显示结果
    plt.figure(figsize=(6, 6))
    plt.imshow(image)
    plt.title(f'预测: {class_names[predicted[0]]}\n置信度: {confidence[0]:.2%}')
    plt.axis('off')
    plt.tight_layout()
    plt.savefig('prediction_result.png', dpi=150, bbox_inches='tight')
    plt.show()

    print(f"📷 图片: {os.path.basename(image_path)}")
    print(f"🤖 模型预测: {class_names[predicted[0]]}")
    print(f"🎯 置信度: {confidence[0]:.2%}")

    print("\n📊 所有类别概率:")
    for i, prob in enumerate(probabilities[0]):
        print(f"  {class_names[i]}: {prob:.2%}")

```

```python
def main():
    """主函数"""
    print("=" * 60)
    print("🌸 花卉识别系统 v1.0 (单文件版)")
    print("=" * 60)

    # 检查GPU
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"💻 使用设备: {device}")

    # 1. 加载数据
    print("\n[1/5] 📁 加载数据集...")
    train_loader, test_loader, class_names = load_and_prepare_data('flower_photos')

    if train_loader is None:
        print("❌ 数据加载失败，请确保数据集存在！")
        print("请下载 flower_photos.tgz 并解压到项目目录")
        return

    # 2. 可视化样本
    print("\n[2/5] 📊 可视化样本图片...")
    visualize_samples(train_loader, class_names)

    # 3. 创建模型
    print(f"\n[3/5] 🧠 创建模型 (共 {len(class_names)} 个类别)")
    model = FlowerClassifier(num_classes=len(class_names)).to(device)
    print("模型结构:")
    print(model)

    # 4. 训练和评估模型
    print(f"\n[4/5] 🏋️ 训练模型 (10个epoch)...")
    train_losses, train_accs, test_accs = train_model(
        model, train_loader, test_loader, device, num_epochs=10
    )

    # 最终评估
    final_test_acc = evaluate_model(model, test_loader, device)
    print(f"\n🏁 最终测试精度: {final_test_acc:.2f}%")

    # 保存模型
    torch.save(model.state_dict(), 'flower_classifier.pth')
    print("💾 模型已保存为 'flower_classifier.pth'")

    # 5. 可视化训练过程
    print("\n[5/5] 📈 可视化训练过程...")
    visualize_training(train_losses, train_accs, test_accs)

    # 6. 单张图片预测（可选）
    test_image_path = 'test_flower.jpg'
    if os.path.exists(test_image_path):
        print("\n🔮 单张图片预测演示...")
        predict_single_image(model, test_image_path, class_names, device)
    else:
        print(f"\n💡 提示: 要测试单张图片，请将图片命名为 'test_flower.jpg' 放在项目目录下")
        print("   可以从这里下载示例图片:")
        print("   https://images.unsplash.com/photo-1562690868-60bbe7293e94?w=400&h=400&fit=crop")

    print("\n" + "=" * 60)
    print("🎉 项目完成！生成文件：")
    print("   1. flower_samples.png - 花卉样本展示")
    print("   2. training_curves.png - 训练过程曲线")
    print("   3. flower_classifier.pth - 训练好的模型")
    print("   4. prediction_result.png - 单张图片预测结果（如有测试图片）")
    print("=" * 60)


if __name__ == '__main__':
    main()
```
